```python
duplicates=df[df.duplicated()]
print("Number of duplicate rows: ", len(duplicates))
```

```python
df.drop_duplicates(inplace=True)
```

```python
print("Shape of dataset: ", df.shape)
```

```python
non_dia= df[df['Diabetes_012'] == 0]
pre_dia = df[df['Diabetes_012'] == 1]
dia = df[df['Diabetes_012'] == 2]

pre_dia_os = pre_dia.sample(len(non_dia), replace=True)
dia_os = dia.sample(len(non_dia), replace=True)

df_new = pd.concat([pre_dia_os,dia_os, non_dia], axis=0)
```

```python
print(df_new['Diabetes_012'].value_counts())
print(df_new.shape)
df_new=df_new.reset_index(drop=True)
```

```python
df_new
```

```python
X=df_new.drop(["Diabetes_012"], axis=1)
Y=df_new[["Diabetes_012"]]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.1, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_train=scaler.fit_transform(x_train)
x_test=scaler.transform(x_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
depth=[]
train_score=[]
test_score=[]
for i in range(1, 40):
    depth.append(i)
    model = DecisionTreeClassifier(max_depth=i)
    model.fit(x_train, y_train)
    train_score.append(model.score(x_train, y_train)*100)
    test_score.append(model.score(x_test, y_test)*100)
```

```python
plt.plot(depth,train_score)
plt.plot(depth,test_score)
plt.title('Model Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Max_depth')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```

```python
model = DecisionTreeClassifier(max_depth=27)
model.fit(x_train, y_train)
print("Max_depth = 27")
print("Train score:",model.score(x_train, y_train)*100)
print("Train score:",model.score(x_test, y_test)*100)
```
